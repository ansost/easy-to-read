## Dependency Lengths

```python
import matplotlib.pyplot as plt
import pandas as pd

train = pd.read_csv("../data/metrics/train_basics.csv")
plt.hist(train["max_dep_length"], bins=100)
plt.xlabel("max_dependency_length")
plt.ylabel("count")
plt.show()
```

## Genre

```python
import matplotlib.pyplot as plt
import pandas as pd

train = pd.read_csv("../data/metrics/train_basics.csv")
genre = train["genre"].str.split("|", expand=True)
genre = genre.stack().value_counts()
genre.plot(kind="bar")
plt.xlabel("genre")
plt.ylabel("count")
plt.show()
```

## Topic

```python
import matplotlib.pyplot as plt
import pandas as pd

train = pd.read_csv("../data/metrics/train_basics.csv")
plt.hist(train["topic"], bins=100)
# tilt x-axis labels
plt.xticks(rotation=90)
plt.xlabel("topic")
plt.ylabel("count")
plt.show()
```
