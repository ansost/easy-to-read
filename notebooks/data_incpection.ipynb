```python
import pandas as pd
from matplotlib import pyplot as plt 
%matplotlib inline

df = pd.read_csv ('../data/trial.csv')

df.head()
```

```python
total = df['num_statements'].value_counts().values.sum()

def fmt(x):
    return '{:.0f}'.format(total*x/100)

plt.pie(df['num_statements'].value_counts().values, labels=df['num_statements'].value_counts().index,autopct=fmt, startangle=90)
plt.title("num_statements")
plt.show()
```
